## Exercice 4

```python
def lecture(fichier):
    with open(fichier,"r", encoding="utf-8") as contenu:
        char = ""
        for ligne in contenu.readlines():
            char += ligne
        return char
```

```python
def ecriture(fichier, contenu):
    with open(fichier,"a", encoding="utf-8", newline="\n") as fp:
        fp.write(contenu)
    return fp
```

```python
def concat(*fichiers):
    with open("fusion.txt", "w", encoding="utf-8") as fusion:
        for fp in fichiers:
            ecriture("fusion.txt",lecture(fp))
```
